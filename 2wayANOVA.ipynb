```
import pandas as pd
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```
data=pd.read_excel('/content/drive/My Drive/Colab Notebooks/Twoway.xlsx')
data
```

```
model=ols('Value ~(college)+(prep_pro)+(college):(prep_pro)',data).fit()
anova_table=sm.stats.anova_lm(model,typ=2)
anova_table
```
